```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Matplotlib basics
x = np.linspace(0,5,11)
x
y = x **3
y
```

```python
plt.plot(x,y)
plt.title("Home")
plt.xlabel("ram")
plt.ylabel("sita")
```

```python
# Subplots in matplotlib
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
# object oriented way
```

```python
fig = plt.figure(figsize=(12,8),dpi=100)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.2])
axis2.plot(x,y)
```

```python
# Types of plot
```

```python
plt.scatter(x,y)
```

```python
plt.boxplot(x)
```

```python
#Saving plots
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Basicplot.png")
```
